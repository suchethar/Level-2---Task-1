````
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
````

````
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# (Change the file name to your dataset path, e.g. "zomato.csv")
df = pd.read_csv("/content/Dataset .csv")

# 1️⃣ Check shape of dataset
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

# 2️⃣ Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Handle missing values (example handling)
# Fill numeric columns with mean, categorical with mode
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)

# 3️⃣ Check and convert data types if necessary
print("\nData types before conversion:")
print(df.dtypes)

# Example: convert numeric columns stored as strings
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col])
        except:
            pass  # ignore if conversion fails

print("\nData types after conversion:")
print(df.dtypes)

# 4️⃣ Analyze target variable distribution ("Aggregate rating")
if "Aggregate rating" in df.columns:
    plt.figure(figsize=(8,5))
    sns.countplot(x='Aggregate rating', data=df, palette='viridis')
    plt.title("Distribution of Aggregate Rating")
    plt.show()

    # Check for class imbalance
    print("\nAggregate rating value counts:")
    print(df['Aggregate rating'].value_counts())
else:
    print("\nColumn 'Aggregate rating' not found in dataset!")
````

````
# 📊 Descriptive Analysis Code

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (change filename if needed)
df = pd.read_csv("/content/Dataset .csv")

# 1️⃣ Basic statistical measures for numerical columns
print("=== Basic Statistical Measures ===")
print(df.describe())

# Calculate additional stats
print("\nMedian values:")
print(df.median(numeric_only=True))

print("\nStandard deviation:")
print(df.std(numeric_only=True))

# 2️⃣ Distribution of categorical variables
categorical_cols = ['Country Code', 'City', 'Cuisines']

for col in categorical_cols:
    if col in df.columns:
        print(f"\n=== Distribution of {col} ===")
        print(df[col].value_counts().head(10))  # top 10 values
        plt.figure(figsize=(8,5))
        sns.countplot(y=col, data=df, order=df[col].value_counts().index[:10], palette="cool")
        plt.title(f"Top 10 {col}")
        plt.show()
    else:
        print(f"\nColumn '{col}' not found in dataset.")

# 3️⃣ Identify top cuisines and cities with the highest number of restaurants
if 'Cuisines' in df.columns:
    top_cuisines = df['Cuisines'].value_counts().head(10)
    print("\n=== Top 10 Cuisines ===")
    print(top_cuisines)
    plt.figure(figsize=(8,5))
    sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="mako")
    plt.title("Top 10 Cuisines with Most Restaurants")
    plt.xlabel("Number of Restaurants")
    plt.ylabel("Cuisine Type")
    plt.show()

if 'City' in df.columns:
    top_cities = df['City'].value_counts().head(10)
    print("\n=== Top 10 Cities ===")
    print(top_cities)
    plt.figure(figsize=(8,5))
    sns.barplot(x=top_cities.values, y=top_cities.index, palette="rocket")
    plt.title("Top 10 Cities with Most Restaurants")
    plt.xlabel("Number of Restaurants")
    plt.ylabel("City")
    plt.show()
````

````
# 🌍 Geospatial Analysis Code

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium  # for interactive maps

# Load dataset (change filename if needed)
df = pd.read_csv("/content/Dataset .csv")

# 1️⃣ Basic check for required columns
required_cols = ['Latitude', 'Longitude', 'City', 'Country Code', 'Aggregate rating']
for col in required_cols:
    if col not in df.columns:
        print(f"⚠️ Column '{col}' not found in dataset!")

# Drop rows with missing latitude or longitude
df = df.dropna(subset=['Latitude', 'Longitude'])

# 2️⃣ Visualize restaurant locations on a world map (using Folium)
# Create base map centered around average location
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=2)

# Add restaurant markers (limited to 500 for better performance)
for _, row in df.head(500).iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=2,
        color='blue',
        fill=True,
        fill_opacity=0.5
    ).add_to(m)

# Display map (in Jupyter/Colab, just use "m" on last line)
m.save("restaurant_locations.html")
print("✅ Map saved as 'restaurant_locations.html' (open it in browser)")

# 3️⃣ Distribution of restaurants across cities
if 'City' in df.columns:
    top_cities = df['City'].value_counts().head(10)
    print("\n=== Top 10 Cities with Most Restaurants ===")
    print(top_cities)

    plt.figure(figsize=(8,5))
    sns.barplot(x=top_cities.values, y=top_cities.index, palette="viridis")
    plt.title("Top 10 Cities with Most Restaurants")
    plt.xlabel("Number of Restaurants")
    plt.ylabel("City")
    plt.show()

# Distribution of restaurants across countries
if 'Country Code' in df.columns:
    top_countries = df['Country Code'].value_counts().head(10)
    print("\n=== Top 10 Countries with Most Restaurants ===")
    print(top_countries)

    plt.figure(figsize=(8,5))
    sns.barplot(x=top_countries.values, y=top_countries.index, palette="magma")
    plt.title("Top 10 Countries with Most Restaurants")
    plt.xlabel("Number of Restaurants")
    plt.ylabel("Country Code")
    plt.show()

# 4️⃣ Correlation between location (latitude/longitude) and rating
if 'Aggregate rating' in df.columns:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x='Longitude', y='Latitude', hue='Aggregate rating', data=df, palette='coolwarm', s=30)
    plt.title("Restaurant Locations Colored by Rating")
    plt.show()

    # Check numeric correlation
    corr = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
    print("\n=== Correlation between location and rating ===")
    print(corr)
else:
    print("\n⚠️ Column 'Aggregate rating' not found, cannot analyze correlation.")
````

````
# 🍽️ Table Booking & Online Delivery Analysis

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (change filename if needed)
df = pd.read_csv("/content/Dataset .csv")

# Check for required columns
required_cols = ['Has Table booking', 'Has Online delivery', 'Aggregate rating', 'Price range']
for col in required_cols:
    if col not in df.columns:
        print(f"⚠️ Column '{col}' not found in dataset!")

# 1️⃣ Percentage of restaurants offering Table Booking and Online Delivery
if 'Has Table booking' in df.columns:
    table_booking_percent = (df['Has Table booking'].value_counts(normalize=True) * 100).round(2)
    print("\n=== Percentage of Restaurants Offering Table Booking ===")
    print(table_booking_percent)

    plt.figure(figsize=(5,5))
    plt.pie(table_booking_percent, labels=table_booking_percent.index, autopct='%1.1f%%', colors=['#66b3ff','#99ff99'])
    plt.title("Table Booking Availability")
    plt.show()

if 'Has Online delivery' in df.columns:
    online_delivery_percent = (df['Has Online delivery'].value_counts(normalize=True) * 100).round(2)
    print("\n=== Percentage of Restaurants Offering Online Delivery ===")
    print(online_delivery_percent)

    plt.figure(figsize=(5,5))
    plt.pie(online_delivery_percent, labels=online_delivery_percent.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
    plt.title("Online Delivery Availability")
    plt.show()

# 2️⃣ Compare average ratings: With vs Without Table Booking
if 'Has Table booking' in df.columns and 'Aggregate rating' in df.columns:
    avg_ratings_booking = df.groupby('Has Table booking')['Aggregate rating'].mean()
    print("\n=== Average Ratings: With vs Without Table Booking ===")
    print(avg_ratings_booking)

    plt.figure(figsize=(6,5))
    sns.barplot(x=avg_ratings_booking.index, y=avg_ratings_booking.values, palette="viridis")
    plt.title("Average Ratings by Table Booking Availability")
    plt.xlabel("Has Table Booking")
    plt.ylabel("Average Rating")
    plt.show()

# 3️⃣ Analyze Online Delivery across different Price Ranges
if 'Has Online delivery' in df.columns and 'Price range' in df.columns:
    plt.figure(figsize=(7,5))
    sns.countplot(x='Price range', hue='Has Online delivery', data=df, palette="coolwarm")
    plt.title("Online Delivery Availability Across Price Ranges")
    plt.xlabel("Price Range")
    plt.ylabel("Number of Restaurants")
    plt.legend(title="Has Online Delivery")
    plt.show()

    # Calculate percentage by price range
    delivery_by_price = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100
    print("\n=== Percentage of Online Delivery by Price Range ===")
    print(delivery_by_price.round(2))
````
